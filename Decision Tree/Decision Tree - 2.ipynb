# Decision Tree Classifer - Iris Dataset

### Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Loading the dataset

```python
iris = pd.read_csv("Iris.csv")
iris
```

```python
iris.isnull().sum()
```

```python
iris.drop(columns=["Id"], inplace=True)
iris
```

### Splitting the Training & Testing Data

```python
X = iris.iloc[:, :-1]
y = iris.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

### Fit & Predict

```python
# Creating Decision Tree Object
clf = DecisionTreeClassifier()
```

```python
# Training the classifier
clf.fit(X_train, y_train)
```

```python
# predicting the y labels for testing data
predictions = clf.predict(X_test)
predictions
```

### Performance Evaluation

```python
acc_score = accuracy_score(y_test, predictions)
print(f"Accuracy Score: {acc_score}")
```

```python
classification_rep = classification_report(y_test, predictions)
print(f"Classification Report: \n {classification_rep}")
```

## Overfitting

```python
#evaluate decision tree performance on train and test sets with different tree depths
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
#synthetic classification dataset
X2, y2 = make_classification(n_samples=10000, n_features=20, n_informative=5, n_redundant=15, random_state=7)
print(X2.shape, y2.shape)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2)
```

```python
train_scores = []
test_scores = []
for i in range(1, 21):
    clf3 = DecisionTreeClassifier(max_depth = i)
    clf3.fit(X2_train, y2_train)
    y_pred_train = clf3.predict(X2_train)
    predictions3 = clf3.predict(X2_test)
    train_score = accuracy_score(y2_train, y_pred_train)
    test_score = accuracy_score(y2_test, predictions3)
    print(f"Max Depth = {i} Train Score = {train_score} Test Score ={test_score}")
    train_scores.append(train_score)
    test_scores.append(test_score)
```

```python
max_depth = [i for i in range(1, 21)]
plt.plot(max_depth, train_scores, '-o', label="Train")
plt.plot(max_depth, test_scores, '-o', label="Test")
plt.legend()
plt.show()
```

The testing accuracy begins to fall somewhere between max depth = 8 and max depth = 9, that is, the classifier is overfitting after max depth = 8
